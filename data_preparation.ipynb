```python
import pandas as pd
```

```python
df = pd.read_csv("loans_full_schema.csv")
```

```python
df.head()
```

```python
df.info
```

```python
customers = df[[
'emp_title',
'emp_length',
'state',
'homeownership',
'annual_income',
'verified_income',
]].drop_duplicates()
```

```python
customers['customer_id'] = range(1, len(customers)+1)
```

```python
df = df.merge(customers, on= [
'emp_title',
'emp_length',
'state',
'homeownership',
'annual_income',
'verified_income',
])
```

```python
customers.head()
```

```python
loans = df[[
'loan_amount',
'term',
'interest_rate',
'installment',
'grade',
'sub_grade',
'loan_purpose',
'issue_month',
'customer_id',
]].copy()
```

```python
loans['loan_id'] = range(1, len(loans)+1)
```

```python
loans.head()
```

```python
df.columns
```

```python
credit_profile = df[[
'customer_id',
'total_credit_lines',
'open_credit_lines',
'total_credit_limit',
'delinq_2y',
'inquiries_last_12m',
'tax_liens',
'public_record_bankrupt']].drop_duplicates().copy()
```

```python
credit_profile.head()
```

```python
payments = df[[
'balance',
'paid_total',
'paid_principal',
'paid_interest',
'paid_late_fees']].copy()
```

```python
payments['payment_id'] = range(1, len(payments)+1)
```

```python
customers.to_csv("customers.csv", index=False)
loans.to_csv("loans.csv", index=False)
credit_profile.to_csv("credit_profile.csv", index=False)
payments.to_csv("payments.csv", index=False)
```
